## Problem:

In this assignment students have to compress racoon grey scale image into
5 clusters. 
In the end, visualize both raw and compressed image and look
for quality difference.

```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
```

```python
img = misc.face(gray = True)
img.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.gray()
plt.figure()
plt.imshow(img)
plt.show()
```

```python
clus_model = cluster.KMeans(n_clusters = 5, n_init = 6 )
clus_model.fit(img.reshape((-1,1)))
```

```python
clus_model.cluster_centers_
```

```python
plt.gray()
plt.figure()
plt.imshow(clus_model.cluster_centers_[clus_model.labels_].reshape((768,1024)))
plt.show()
```
