1. In this assignment students have to transform iris data into
   3-dimension and plot 3-d chart with transformed dimension and
   colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
data = datasets.load_iris()
data
```

```python
data.feature_names
```

```python
x = data.data
y = data.target
```

```python
pca = decomposition.PCA(n_components = 3)
transformed_x = pca.fit_transform(x)
```

```python
fig = plt.figure(figsize = (10,8))
ax = Axes3D(fig, azim = 110, elev = 30)
ax.scatter(transformed_x[:,0],transformed_x[:,1],transformed_x[:,2],
           c = y )

ax.set(title = 'IRIS DATASET',
       xlabel = 'X-axis', 
       ylabel = 'Y-axis', 
       zlabel = 'Z-axis',
       )

for name,label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(transformed_x[y == label,0].mean() ,
              transformed_x[y == label, 1].mean() + 1.7,
              transformed_x[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=1, edgecolor='w', facecolor='w'))

plt.show()
```
